```python
import matplotlib.pyplot as plt
import numpy as np
categories=['female','male','other']
a=df['gender'].value_counts()
print(a)
values_bar=a
plt.figure(figsize=(7,6))
plt.bar(categories,a,color=['red','blue','green'])
plt.title('bar chart')
plt.xlabel("categories")
plt.ylabel("gender")
plt.show()
```

```python
categories=['female','male','other']
y=df['gender'].value_counts()
plt.figure(figsize=(7,3))
plt.scatter(categories,y,color='green',marker='o')
plt.title('scatter plot')
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.show()
```
